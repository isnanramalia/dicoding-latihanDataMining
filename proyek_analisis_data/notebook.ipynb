CLEANING DATA

```python
# membersihkan data customer_df
customers_df.drop_duplicates(inplace=True) # menghapus duplikasi data
print("Jumlah duplikasi: ", customers_df.duplicated().sum())

customers_df[customers_df.gender.isna()] # menangani missing values
customers_df.gender.value_counts() # menangani missing values
```

```python
customers_df.fillna(value="Prefer not to say", inplace=True) # mengisi missing values
customers_df.isna().sum() # mengecek missing values
```

```python
customers_df[customers_df.age == customers_df.age.max()] # menangani outliers
```

```python
customers_df.age.replace(customers_df.age.max(), 70, inplace=True) # mengganti outliers
```

```python
customers_df[customers_df.age == customers_df.age.max()] # mengecek outliers
```

```python
customers_df.age.replace(customers_df.age.max(), 50, inplace=True)
customers_df.describe() 
```
